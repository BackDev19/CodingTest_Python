```python
def solution(n, k, cmd):
    
    cur = k
    node = {i : [i - 1, i + 1] for i in range(n)}
    node[0] = [None, 1]
    node[n-1] = [n-2, None]
    stack = []
    answer = ['O'] * n
    
    for c in cmd:
        if c == 'C':
            prev, next = node[cur]
            stack.append([prev, cur, next])
            answer[cur] = 'X'
            
            # 커서 이동, 노드끼리 연결해주기
            if next != None:
                node[next][0] = prev
            if prev != None:
                node[prev][1] = next
            
            # 커서 이동
            if next != None:
                cur = next
            elif prev != None:
                cur = prev
        
        elif c == 'Z':
            prev, remove_idx, next = stack.pop()
            answer[remove_idx] = 'O'
            
            # 이전과 다음의 노드에 삭제한 노드의 idx를 넣어주기
            if next != None:
                node[next][0] = remove_idx
            if prev != None:
                node[prev][1] = remove_idx
        
        # 커서 이동
        else:
            direction, idx = c.split(' ')
            idx = int(idx)
            
            if direction == 'U':
                for _ in range(idx):
                    cur = node[cur][0]
            else:
                for _ in range(idx):
                    cur = node[cur][1]
        
    return ''.join(answer)
```

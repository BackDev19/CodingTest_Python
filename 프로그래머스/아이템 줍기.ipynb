```python
from collections import deque
def solution(rectangle, characterX, characterY, itemX, itemY):
    # 해당 사각형 좌표를 표시
    def drawLine(pos):
        pos = list(map(lambda x : x * 2, pos))
        x1, y1, x2, y2 = pos
 
        for i in range(y1, y2 + 1):
            for j in range(x1, x2 + 1):
                # 내부에 있는 값들은 8으로 표시
                if y1 < i < y2 and x1 < j < x2:
                    board[i][j] = 8
                # 테두리는 1로 표시
                elif board[i][j] != 8:
                    board[i][j] = 1
    
    # 최단 거리 탐색
    def bfs(now_x, now_y, item_x, item_y):
        q = deque([[now_y, now_x]])
        visited = [[-1] * (size) for _ in range(size)]
        visited[now_y][now_x] = 0
        
        dy = [-1, 0, 1, 0]
        dx = [0, 1, 0, -1]
        
        while q:
            y, x = q.popleft()
            if x == item_x and y == item_y:
                return visited[y][x]
            
            for i in range(4):
                nx = x + dx[i]
                ny = y + dy[i]
                
                if 0 <= nx < size and 0 <= ny < size:
                    if visited[ny][nx] == -1 and board[ny][nx] == 1:
                        visited[ny][nx] = visited[y][x] + 1
                        q.append([ny, nx])
        
    
    # 50 이하이므로, 51 * 2배
    size = 102
    board = [[9] * (size) for _ in range(size)]
    
    # 선 그려주기. 단, 2배로 그려줘야 한칸이동하는 부분이 표시가 난다
    for rect in rectangle:
        drawLine(rect)
    
    # character, item 위치도 * 2
    # answer는 board를 2배 해서 거리를 구했기 때문에 // 2
    answer = bfs(characterX * 2, characterY * 2, itemX * 2, itemY * 2) // 2
    
    return answer
```

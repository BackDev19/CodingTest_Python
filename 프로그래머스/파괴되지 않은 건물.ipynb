```python
# 내구도 감소, 0 이하 파괴
# 아군은 내구도 증가, 건물 복구
# 내구도는 계속 감소함

# [type, r1, c1, r2, c2, degree]
def solution(board, skill):
    answer = 0
    
    n, m = len(board), len(board[0])
    table = [[0] * (m + 1) for _ in range(n + 1)]
    
    # 누적할 변수 세팅
    def doSkill(skill):
        nonlocal table
        type, r1, c1, r2, c2, degree = skill
        # 공격인 경우, degree를 음수로
        if type == 1:
            degree = -degree
        
        # 각 누적합이 0이 될 수 있도록 세팅
        table[r1][c1] += degree
        table[r1][c2 + 1] -= degree
        table[r2 + 1][c1] -= degree
        table[r2 + 1][c2 + 1] += degree
    
    # 누적합 진행
    def accumulateSum():
        # 1. 모든행에서 오른쪽으로 누적
        for i in range(n + 1):
            for j in range(1, m + 1):
                table[i][j] += table[i][j-1]
                
        # 2. 모든 열에서 아래쪽으로 누적
        for i in range(m + 1):
            for j in range(1, n + 1):
                table[j][i] += table[j-1][i]
        
    # 스킬 사용. 누적할 변수 세팅
    for x in skill:
        doSkill(x)
    
    # 누적 합
    accumulateSum()
    
    # board와 누적 테이블 연산
    for i in range(n):
        for j in range(m):
            board[i][j] += table[i][j]
            if board[i][j] > 0:
                answer += 1
    
    return answer
```

```python
import sys
sys.setrecursionlimit(10**6)
def solution(nodeinfo):

    for i in range(len(nodeinfo)):
        nodeinfo[i].append(i+1)
    
    sorted_nodes = sorted(nodeinfo, key = lambda x : (-x[1], x[0]))
    
    answer = []
    answer.append(preOrder(sorted_nodes[0], sorted_nodes))
    answer.append(postOrder(sorted_nodes[0], sorted_nodes))
    
    return answer

def preOrder(now, nodes):
    result = [now[2]]
    
    left_nodes = []
    right_nodes = []
    
    for i in range(len(nodes)):
        if nodes[i][0] < now[0]:
            left_nodes.append(nodes[i])
        elif nodes[i][0] > now[0]:
            right_nodes.append(nodes[i])
    
    if len(left_nodes) > 0:
        result += preOrder(left_nodes[0], left_nodes)
    if len(right_nodes) > 0:
        result += preOrder(right_nodes[0], right_nodes)
    
    return result

def postOrder(now, nodes):
    result = []
    
    left_nodes = []
    right_nodes = []
    
    for i in range(len(nodes)):
        if nodes[i][0] < now[0]:
            left_nodes.append(nodes[i])
        elif nodes[i][0] > now[0]:
            right_nodes.append(nodes[i])
    
    if len(left_nodes) > 0:
        result = postOrder(left_nodes[0], left_nodes) + result
    if len(right_nodes) > 0:
        result = result + postOrder(right_nodes[0], right_nodes)
    
    result += [now[2]]
    
    return result
```

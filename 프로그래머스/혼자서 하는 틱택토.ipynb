```python
def solution(board):
    # 완성이 되었는지
    def isComplete(shape):
        # 행, 열 체크
        for i in range(3):
            if board[i][0] == board[i][1] == board[i][2] == shape:
                return True
            if board[0][i] == board[1][i] == board[2][i] == shape:
                return True
        
        # 좌우 대각선 체크
        if board[0][0] == board[1][1] == board[2][2] == shape:
            return True
        # 우상 좌하 체크
        if board[0][2] == board[1][1] == board[2][0] == shape:
            return True
        
        return False
    
    def check():
        # O와 X가 동시에 완성될 수 없음
        if o_complete and x_complete:
            return False
        
        # 완성되지 않은 경우 O - 1 <= X <= O
        if not (o_complete and x_complete) and not (o_cnt - 1 <= x_cnt <= o_cnt):
            return False
        
        # O가 완성된 경우 X = O-1
        if (o_complete and not x_complete) and not (x_cnt == o_cnt - 1):
            return False
            
        # X가 완성된 경우 X = O
        if (not o_complete and x_complete) and not (x_cnt == o_cnt):
            return False
            
        # 나올 수 있는 상황 1    
        return True
        
    answer = -1
    # o와 x의 개수
    o_cnt = 0
    x_cnt = 0
    for i in range(3):
        o_cnt += board[i].count('O')
        x_cnt += board[i].count('X')
    
    # O와 X가 완성이 되었는지 확인
    o_complete = isComplete('O')
    x_complete = isComplete('X')
    
    answer = 1 if check() else 0
    
    return answer
```

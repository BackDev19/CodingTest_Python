```python
import sys
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# 크기가 R x C
# 양은 가만히, 늑대는 인접 칸을 이동 가능
# 두 칸이 변을 공유한 경우 인접 => 상, 하, 좌, 우

# 목장에 울타리를 설치하여, 늑대가 양이 있는 칸으로 갈 수 없게 한다

# 입력
# R, C
# R개의 줄에 목장의 상태. s는 양, w는 늑대

# 양이 있는 칸으로 갈 수 없게 할 수 있다면 1 갈 수 있다면 0
# 울타리는 D

# 갈 수 잇는 경우 -> 상하좌우가 붙어잇는 경우

r, c = map(int, input().split())
graph = [list(input().rstrip()) for _ in range(r)]

dx = [1, 0, -1, 0]
dy = [0, 1, 0, -1]

answer = False

for i in range(r):
    for j in range(c):
        # 해당 위치가 늑대라면 상하좌우를 탐색
        if graph[i][j] == 'W':
            for k in range(4):
                nx = i + dx[k]
                ny = j + dy[k]
                if 0 <= nx < r and 0 <= ny < c:
                    # 양을 발견했다면 0
                    if graph[nx][ny] == 'S':
                        answer = True
            
        # 해당 위치가 늑대 or 양이 아니라면, 울타리 설치
        if graph[i][j] == '.':
            graph[i][j] = 'D'

# 양을 발견했다면
if answer:
    print(0)
# 양을 발견하지 못했다면 1
else:
    print(1)
    for i in range(r):
        for j in range(c):
            print(graph[i][j], end = '')
        print()

```

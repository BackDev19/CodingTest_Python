```python
import sys
import copy
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# 가로줄과 세로줄에는 1부터 9까지의 숫자가 한 번씩만
# 3 x 3 정사각형 안에도 1부터 9까지의 숫자가 한 번씩만

# for ~ 9, j = ~ 9
# 행, 열, 3 x 3 check 함수로, visited == False인 숫자들 찾기
# visited False인 idx들 순차적으로 backtracking 진행
# check == False면 return

# 1. check함수로 방문되지 않은 visited 찾기
# 2. 하나씩 visited = True, False해주면서, 다음 0인 위치 dfs() 진행

graph = [list(map(int, input().split())) for _ in range(9)]

zero = []

for i in range(9):
    for j in range(9):
        if graph[i][j] == 0:
            zero.append((i, j))

def check(x, y, visited):
    # 행, 열 check
    # 0인 부분은 continue
    for i in range(9):
        if graph[x][i] == 0:
            continue
        visited[graph[x][i]] = True
    
    for i in range(9):
        if graph[i][y] == 0:
            continue
        visited[graph[i][y]] = True
    
    # 3 x 3 배열의 숫자 확인
    # 위치 확인
    for i in range(x // 3 * 3, x // 3 * 3 + 3):
        for j in range(y // 3 * 3, y // 3 * 3 + 3):
            if graph[i][j] == 0:
                continue
            visited[graph[i][j]] = True

answer = []
def dfs(L):

    # 모두 돌았는데, 비어있는 숫자가 있다면
    if L == len(zero):
        for i, j in zero:
            if graph[i][j] == 0:
                return
        # 비어있는 숫자가 없다면
        global answer
        answer = copy.deepcopy(graph)
        return

    now_x, now_y = zero[L]
    visited = [False] * 10
    # 하나씩 check 처리
    check(now_x, now_y, visited)

    # 방문하지 않았다면, 해당 위치를 True로 만들고 visited
    for i in range(1, 10):
        if not visited[i]:
            # 방문했던 idx에 번호를 넣어줄 것
            graph[now_x][now_y] = i
            dfs(L + 1)
            graph[now_x][now_y] = 0

    return

dfs(0)
# print(*graph, sep = '\n')
for x in answer:
    print(*x, sep = ' ')
```

```python
import sys
from collections import deque
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# N x M의 행렬로 표현되는 맵
# 맵에서 0은 이동할 수 있는 곳
# 1은 이동할 수 없는 벽

# 1, 1에서 N, M의 위치까지 이동
# 최단거리로 이동
# 맵에서 가장 적은 개수의 칸을 지나는 경로
# 시작하는 칸과 끝나는 칸도 포함

# 벽 한개까지는 부술 수 있음
# 최단 경로 구하기

# N, M.
# N개의 줄에 M개의 숫자로 맵이 주어진다
# (1, 1) (N, M)은 항상 0

n, m = map(int, input().split())
graph = [list(map(int, input().rstrip())) for _ in range(n)]
visit = [[[0] * (m) for _ in range(n)] for _ in range(2)]

def bfs():
    q = deque()
    q.append((0, 0, 0))
    visit[0][0][0] = 1
    visit[1][0][0] = 1

    dx = [-1, 0, 1, 0]
    dy = [0, 1, 0, -1]

    while q:
        x, y, flag = q.popleft()
        if x == n-1 and y == m-1:
            return visit[flag][x][y]

        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx < n and 0 <= ny < m:
                # 벽인 경우, 벽을 뚫지 않은 얘가 현재 위치를 방문했어야 함
                if graph[nx][ny] == 1 and flag == 0:
                    visit[1][nx][ny] = visit[flag][x][y] + 1
                    # 벽을 않은 노드에 + 1 더해주기
                    q.append((nx, ny, 1))
                
                # 벽이 아닌 경우, 벽을 방문한 경우
                elif graph[nx][ny] == 0 and not visit[flag][nx][ny]:
                    visit[flag][nx][ny] = visit[flag][x][y] + 1
                    q.append((nx, ny, flag))
                    

answer = bfs()
if not answer:
    answer = -1

# print(*visit[0], sep='\n')
# print()
# print(*visit[1], sep='\n')
print(answer)
```

```python
import sys
from collections import deque
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# 눈금의 간격이 1인 M x N 좌표는 (N, M)
# 눈금에 맞추어 K개의 직사각형
# K개의 직사각형의 내부를 제외한 나머지 부분이 몇 개의 분리된 영역으로 나누어진다

# ex) M = 5, N = 7
# 분리된 영역의 넓이가 얼마인지

# 입력
# M, N, K가 주어진다
# K개의 줄에는 왼쪽 아래, 오른쪽 위 x, y가 주어짐
# 왼쪽 아래 (0,0) 에서부터 오른쪽 위 (N, M)


# 0 2 4 4 -> (x, y) -> (0, 2), (4, 4)

# 1. for i in range(x1, x2 - x1 + 1)
# 2. for j in range(y1, y2 - y1 + 1) 까지 1로 채워주기
# 3. 0인 공간들의 개수 count

m, n, k = map(int, input().split())
graph = [[0] * (n) for _ in range(m)]
visited = [[False] * (n+1) for _ in range(m+1)]

for _ in range(k):
    x1, y1, x2, y2 = map(int, input().split())

    for i in range(y1, y2):
        for j in range(x1, x2):
            graph[i][j] = 1

def bfs(i, j):
    q = deque()
    q.append((i, j))
    visited[i][j] = True

    dx = [1, 0, -1, 0]
    dy = [0, 1, 0, -1]

    count = 1

    while q:
        x, y = q.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            # nx는 0 ~ 4까지 -> m, ny는 0 ~ 6까지 -> n 
            if 0 <= nx < m and 0 <= ny < n :
                if not visited[nx][ny] and graph[nx][ny] == 0:
                    visited[nx][ny] = True
                    q.append((nx, ny))
                    count += 1

    return count

answer = []
# x좌표인 n, y좌표인 m
for i in range(m):
    for j in range(n):
        if not visited[i][j] and graph[i][j] == 0:
            answer.append(bfs(i, j))

answer.sort()
print(len(answer))
for x in answer:
    print(x, end = ' ')

```

```python
# 엄지손가락은 상하좌우 4가지 방향으로만.
# 키패드 이동 한 칸은 거리로 1

# 147 왼쪽
# 369 오른쪽
# 2580 가까운 엄지손가락
# 거리가 같다면, 손잡이별로

# 1. 딕셔너리로 위치 값 계산
# 2. 147이면 왼쪽, 369이면 오른쪽, 나머지는 거리계산

def touch(number, hand):
    
    global left, right
    
    # 왼쪽 터치인 경우
    if number in [1, 4, 7]:
        left = keypad[number]
        return 'L'
    
    # 오른쪽 터치인 경우
    elif number in [3, 6, 9]:
        right = keypad[number]
        return 'R'
    
    # 나머지 keypad
    else:
        result = getDistance(left, keypad[number]) - getDistance(right, keypad[number])
        # result가 음수인 경우 -> right의 거리가 더 먼 것
        if result < 0:
            left = keypad[number]
            return 'L'
        # result가 양수인 경우 -> left의 거리가 더 먼 것
        elif result > 0:
            right = keypad[number]
            return 'R'
        # result가 0인 경우 손잡이로 이동시켜주기
        else:
            if hand == 'right':
                right = keypad[number]
                return 'R'
            else:
                left = keypad[number]
                return 'L'
        

# 손과 keypad의 거리 구하기
def getDistance(hand_pos, key_pos):
    distance = 0
    # 행과 열을 각각 뺀 뒤에 더해주기
    distance += abs(hand_pos[0] - key_pos[0])
    distance += abs(hand_pos[1] - key_pos[1])
    
    return distance
        
def solution(numbers, hand):
    answer = ''
    
    global keypad, left, right
    
    # keypad 숫자별 위치 등록
    keypad = dict()
    for i in range(3):
        for j in range(3):
            keypad[3*i+j+1] = [i, j]
    keypad[0] = [3, 1]
    
    # 손가락 위치 지정
    left = [3, 0]; right = [3, 2]
    
    for i in range(len(numbers)):
        answer += touch(numbers[i], hand)
    
    
    return answer
```

```python
# 두 점수가 모두 낮은 경우가 한번이라도 있다면 그 사원은 인센티브 x
# 그렇지 않은 경우 두 점수의 합이 높은 순으로 석차를 내어 석차에 따라 인센티브 차등 지급

# 이때, 두 점수의 합이 동일한 사원들은 동석차, 동석차의 수만큼 다음 석차는 건너 뛴다.
# ex) 1등이 2명이라면 다음은 3등

# 근무 태도 점수와 동료 평가 점수 목록 scores, 완호의 석차를 return

# 5번째 사원은 3 또는 4번째 사원보다 근무 태도 점수와 동료 평가 점수가 모두 낮기 때문에 인센티브를 받을 수 없다
# 2, 3, 4 사원은 두 점수의 합이 5점으로 최고점

# 근무태도, 동료평가 점수
# [0]는 완호의 점수

import copy
# sum을 기준으로 정렬
def order_sum(scores):
    q = []
    for idx, score in scores:
        q.append((idx, sum(score)))
    
    # sum을 기준으로 정렬
    q.sort(key = lambda x : -x[1])
    return q

def rank_calc(scores, rank):
    count = 1
    # 첫째 scores의 rank에 1을 넣어줌
    rank[scores[0][0]] = 1
    prev_idx = scores[0][0]
    for i in range(len(scores)):
        # 해당 rank가 -1이라면 continue
        if rank[scores[i][0]] == -1:
            continue
        
        if scores[i][1] < scores[i-1][1]:
            rank[scores[i][0]] = count
        # 값이 동일하다면,
        # 이전의 값이 있던 idx값을 넣어줌
        else:
            rank[scores[i][0]] = rank[prev_idx]
            prev_idx = scores[i][0]
        
        count += 1

def solution(scores):
    answer = 0
    
    order = []
    # 일단, scores 순으로 순서를 저장해주기
    for idx, score in enumerate(scores):
        order.append((idx, score))
    
    # 근무 태도 점수 기준으로 내림차순. 같으면 동료평가를 오름차순
    # -> [4 5], [4, 6]이면, 조건에 안걸리기도 하고, 이미 근무 태도 기준으로는 내림차순이기 때문에 항상 작은 상황
    # 따라서 지금까지의 최댓값을 구한 후, 동료 평가가 최댓값보다 작으면 둘 다 작게되는 것임
    order.sort(key = lambda x : (-x[1][0], x[1][1]))
    
    maxPeer = 0
    rank = [0] * (len(scores))
    # 인사고과를 받지 못하는 rank들 -1로 표시
    for idx, score in order:
        if score[1] < maxPeer:
            rank[idx] = -1
        maxPeer = max(maxPeer, score[1])
    
    # sum을 가장 큰 순으로 정렬
    sum_list = order_sum(order)
    
    # 이제 하나씩 크기 비교해주면서, rank가 -1이라면 continue
    # 크기가 더 작다면, 
    rank_calc(sum_list, rank)
    
    answer = rank[0]
    
    return answer
```

```python
import sys
from collections import deque
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# 실패한 케이스

# 1부터 n까지 서로 다른 번호가 매겨진 등대 n개
# 뱃길 n-1개
# 양쪽 끝 등대 중 적어도 하나는 켜져 있도록 등대를 켜 놓기 -> 선이 연결되어있는 등대 중 하나는 무조건 켜놓기
# 등대 개수의 최솟값을 return

# 1. 자신의 길이와 속해있는 노드들의 개수를 비교
# 2. 자신의 길이가 크다면, 불을 켜주기
def solution(n, lighthouse):
    # 다른 한쪽이 켜져있는지 확인
    def nearLightOn(nodes):
        for node in nodes:
            if isLightOn[node]:
                return True
        
        return False
    
    answer = 0
    nodes = [[] for _ in range(n + 1)]
    for a, b in lighthouse:
        nodes[a].append(b)
        nodes[b].append(a)

    # 등대의 연결 개수
    connected_lighthouse = [len(x) for x in nodes]
    isLightOn = [False] * (len(nodes))

    # 1. 자신의 길이와 속해있는 노드들의 개수를 비교
    for i in range(1, len(nodes)):
        # 근처에 등대가 켜져있다면, continue
        if nearLightOn(nodes[i]):
            continue

        # 2. 자신의 길이가 크다면, 불을 켜주기
        # 자신과 연결된 전구 중 길이가 가장 긴 전구를 켜기
        max_node = i
        for node in nodes[i]:
            if connected_lighthouse[max_node] < connected_lighthouse[node]:
                max_node = node

        isLightOn[max_node] = True

    # print(nodes)
    # print(connected_lighthouse)
    # print(isLightOn)
    answer = isLightOn.count(True)
    return answer

# solution(8, [[1, 2], [1, 3], [1, 4], [1, 5], [5, 6], [5, 7], [5, 8]])
solution(10, [[4, 1], [5, 1], [5, 6], [7, 6], [1, 2], [1, 3], [6, 8], [2, 9], [9, 10]])
```

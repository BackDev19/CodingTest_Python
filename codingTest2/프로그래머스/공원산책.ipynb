```python
import sys
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# 장애물, 범위 밖이라면 명령 무시
# 최종 위치

def solution(park, routes):

    # 장애물이 있다면 True
    def is_obstacle(nx, ny):

        for i in range(1, num + 1):
            nx = pos_x + dx[dir] * i
            ny = pos_y + dy[dir] * i
            if not (0 <= nx < h and 0 <= ny < w) or park[nx][ny] == 'X':
                return True
            
        return False
    
    park = list(map(list, park))
    routes = [route.split(" ") for route in routes]

    # 가로, 세로
    w, h = len(park[0]), len(park)

    dx = [-1, 0, 1, 0]
    dy = [0, 1, 0, -1]

    direction = {
        'N' : 0,
        'E' : 1,
        'S' : 2,
        'W' : 3
    }

    # 시작지점 찾기
    for i in range(h):
        for j in range(w):
            if park[i][j] == 'S':
                pos_x = i
                pos_y = j

    for op, num in routes:
        dir = direction[op]
        num = int(num)

        # 이동가능한 경우
        if not is_obstacle(pos_x, pos_y):
            pos_x = pos_x + dx[dir] * num
            pos_y = pos_y + dy[dir] * num
        
    return pos_x, pos_y

park = 	["OSO", "OOO", "OXO", "OOO"]
routes = ["E 2", "S 3", "W 1"]
solution(park, routes)

```

```python
import sys
from collections import deque
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# 작업 N개
# 걸리는 시간

# 선행 관계
# 선행관계 1이상 k-1 이하
# 선행 관계에 있는 작업이 반드시 하나 이상 존재

# 모든 작업을 완료하기 위해 필요한 최소 시간
global n
n = int(input())

# n + 1, n개의 줄
# 2번째 줄은 1번 작업, 3번째 줄은 2번 작업...
# 해당 작업에 걸리는 시간, 그 작업에 대한 선행 관계에 있는 작업들의 개수
# 작업들의 번호

def topology_sort():
    q = deque()
    for i in range(1, n+1):
        if indegree[i] == 0:
            q.append(i)

    while q:
        now = q.popleft()
        dp[now] += distance[now]

        for x in graph[now]:
            indegree[x] -= 1
            dp[x] = max(dp[x], dp[now])

            if indegree[x] == 0:
                q.append(x)


dp = [0] * (n+1)
distance = [0] * (n+1)
indegree = [0] * (n+1)
graph = [[] for _ in range(n+1)]

for i in range(1, n+1):
    data = list(map(int, input().split()))
    distance[i] += data[0]

    for j in range(data[1]):
        graph[data[2+j]].append(i)
        indegree[i] += 1

# print(distance)
# print(indegree)
# print(*graph, sep='\n')

topology_sort()
print(max(dp))
```

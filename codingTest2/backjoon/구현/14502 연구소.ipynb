```python
import sys
from collections import deque
import copy
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# N x M인 직사각형
# 0 - 빈칸, 1 - 벽, 2 - 바이러스
# 벽을 3개 세운 뒤의 안전 영역의 최댓값 

# 벽세우기
def dfs(L, s):

    # 벽을 3개 세운 경우
    if L == 3:
        global answer
        answer = max(answer, bfs(copy.deepcopy(board)))
        return

    # 벽 세우기
    for i in range(s, len(blank)):
        x ,y = blank[i]
        board[x][y] = 1
        dfs(L + 1, i + 1)
        board[x][y] = 0

# 바이러스
def bfs(board):

    # 바이러스 위치 찾기
    dq = deque()
    for i in range(n):
        for j in range(m):
            if board[i][j] == 2:
                dq.append([i, j])

    dx = [-1, 0, 1, 0]
    dy = [0, 1, 0, -1]

    # 바이러스 퍼뜨리기
    while dq:
        x, y = dq.popleft()

        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if (0 <= nx < n and 0 <= ny < m) and board[nx][ny] == 0:
                board[nx][ny] = 2
                dq.append([nx, ny])


    # 퍼지지 않은 빈 칸 찾기
    result = 0
    for i in range(n):
        result += board[i].count(0)
    
    return result

n, m = map(int, input().split())
board = [list(map(int, input().split())) for _ in range(n)]

answer = 0
blank = []
# 빈 칸 찾기
for i in range(n):
    for j in range(m):
        if board[i][j] == 0:
            blank.append([i, j])

dfs(0, 0)
print(answer)
```

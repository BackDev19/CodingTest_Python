```python
import sys
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# 20207 달력
# 1 .1 ~ 1000까지 배열
# 2. 0으로 끝나기 전에 가장 높은 수를 count
def solution(n):
    answer = 0
    schedule_cnt = [0] * 1001

    # 입력
    for _ in range(n):
        s, e = map(int, input().split())

        # s ~ e까지 일정추가
        for i in range(s, e+1):
            schedule_cnt[i] += 1

    # 면적 출력
    # 0부터 시작 start. 0으로 종료 end
    # 해당 구간 내에서 가장 큰 값 찾기 : max_num
    # sum((end - start) * max_num)
    is_progress = False
    start_idx = -1
    end_idx = -1
    max_num = 0
    for i in range(1, 1001):
        end_idx = i

        # 시작 : 0 이상인 idx
        if not is_progress and schedule_cnt[i] >= 1:
            is_progress = True
            start_idx = i
        
        # 종료 : 0으로 끝나는 idx
        # 1. 면적 계산
        # 2. max_num 초기화
        elif is_progress and schedule_cnt[i] == 0:
            is_progress = False
            answer += max_num * (end_idx - start_idx)

            max_num = 0

        # 스케줄이 있는 경우 중, 가장 큰 수
        if is_progress:
            max_num = max(max_num, schedule_cnt[i])

    # 1000까지 수가 채워진 경우
    if is_progress:
        answer += max_num * (end_idx - start_idx)


    print(answer)


n = int(input())
solution(n)
```

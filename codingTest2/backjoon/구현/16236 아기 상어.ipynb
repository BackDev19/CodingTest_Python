```python
import sys
from collections import deque
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# N x N
# 물고기 M마리와 아기 상어 1마리
# 아기 상어와 물고기는 모두 크기
# 아기 상어 크기는 2. 1초에 상하좌우로 인접한 한 칸씩 이동
# 자신의 크기보다 작은 물고기만 먹을 수 있다.
# 큰 물고기가 있는 칸만 이동 불가

# 더 이상 먹을 수 있는 물고기가 없다면, 도움 요청
# 먹을 수 있는 물고기가 1마리보다 많다면, 거리가 가장 가까운 물고기. 거리는 칸의 최솟값
# 거리가 가장 가까운 물고기가 많다면, 가장 위에. 그러한 물고기가 여러마리라면, 가장 왼쪽

# 자신의 크기가 같은 수의 물고기를 먹을 때마다 크기가 1 증가
# 도움 요청하는데 걸리는 시간

# 물고기 찾기
def bfs(start_x, start_y):
    dq = deque()
    dq.append([start_x, start_y])

    visited = [[-1] * n for _ in range(n)]
    visited[start_x][start_y] = 0

    dx = [-1, 0, 1, 0]
    dy = [0, 1, 0, -1]

    fish = []

    while dq:
        x, y = dq.popleft()

        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if (0 <= nx < n and 0 <= ny < n) and visited[nx][ny] == -1:
                # 지나갈 수 있음
                if board[nx][ny] <= shark_size:
                    visited[nx][ny] = visited[x][y] + 1
                    dq.append([nx, ny])

                # 먹을 수 있음
                if board[nx][ny] != 0 and board[nx][ny] < shark_size:
                    fish.append([visited[nx][ny], nx, ny])
                
    if fish:
        # 물고기가 있는 경우, 거리 > 위 > 왼쪽 순 (하나도 같이 포함)
        fish.sort(key = lambda x : (x[0], x[1], x[2]))
        return fish[0]

    # 주변에 먹을 수 있는 물고기가 없다면 -1
    else:
        return -1

n = int(input())
board = [list(map(int, input().split())) for _ in range(n)]
shark_size = 2
pos_x, pos_y = 0 ,0

# shark 초기 위치 찾기
for i in range(n):
    for j in range(n):
        if board[i][j] == 9:
            pos_x, pos_y = i, j
            board[i][j] = 0

food = 0
answer = 0
while True:
    result = bfs(pos_x, pos_y)
    if result == -1:
        break

    # 물고기를 먹은 경우
    dist, pos_x, pos_y = result
    board[pos_x][pos_y] = 0
    food += 1

    # shark_size만큼 먹었다면, shark_size를 하나 증가
    if shark_size == food:
        shark_size += 1
        food = 0
    
    answer += dist

print(answer)
```

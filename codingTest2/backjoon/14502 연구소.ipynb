```python
import sys
import copy
from itertools import combinations
from collections import deque
input = sys.stdin.readline

sys.stdin = open("input.txt", "rt")

n, m = map(int, input().split())
board = [list(map(int, input().split())) for _ in range(n)]

# 1. combination을 사용
# 2. 3곳을 벽으로 막고 bfs 진행

# virus위치와, 비어있는 공간 파악
virus_pos = []
blank_pos = []

def bfs(board):
    q = deque()
    for virus in virus_pos:
        x, y = virus
        q.append([x, y])

    dx = [-1, 0, 1, 0]
    dy = [0, 1, 0, -1]

    idx = 0

    while q:
        x, y = q.popleft()

        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
                # 주변이 빈 공간이라면, 바이러스 감염 + 추가
            if 0 <= nx < n and 0 <= ny < m and board[nx][ny] == 0:
                board[nx][ny] = 2
                q.append([nx, ny])

    
    result = 0
    for i in range(n):
        result += board[i].count(0)

    return result

for i in range(n):
    for j in range(m):
        if board[i][j] == 2:
            virus_pos.append([i, j])
        elif board[i][j] == 0:
            blank_pos.append([i, j])

answer = 0

for wall_pos in combinations(blank_pos, 3):
    board_copy = copy.deepcopy(board)
    # 벽 세우기
    for wall in wall_pos:
        x, y = wall
        board_copy[x][y] = 1

    answer = max(answer, bfs(board_copy))

print(answer)
```

```python
import sys
sys.stdin = open("input.txt", "rt")
input = sys.stdin.readline

# 집의 크기가 R x C
# r, c에 있는 미세먼지의 양을 실시간으로 모니터링

# 공기청정기가 1행 2열을 차지
# 미세먼지의 양 Ar, Ac

# 미세먼지 모든 칸에서 확산
# 칸이 없거나 공기청정기 -> 확산 x
# Ar,c / 5. 소수점은 버린다
# 원래의 미세먼지는 Ar,c - Ar,c / 5 x 확산된 방향의 개수

# 공기청정기 작동
# 위쪽은 반시계방향
# 아래쪽은 시계방향으로 먼지가 이동

# 바람이 불면 미세먼지가 바람의 방향대로 모두 한 칸씩 이동한다
# 공기청정기로 들어간 미셈너지는 모두 정화

global r, c, t
r, c, t = map(int, input().split())
global graph
graph = [list(map(int, input().split())) for _ in range(r)]

dirty = []
airclearner = []

dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]
def expand():
    global graph
    new = [[0] * c for _ in range(r)]

    for i in range(r):
        for j in range(c):
            if graph[i][j] == -1:
                new[i][j] = -1
                continue

            # 먼지인 경우
            if graph[i][j] != -1 and graph[i][j] != 0:
                count = 0
                for k in range(4):
                    nx = i + dx[k]
                    ny = j + dy[k]
                    # 범위 안에 있고, 공기청정기가 아닌 경우
                    if 0 <= nx < r and 0 <= ny < c and graph[nx][ny] != -1:
                        new[nx][ny] += (graph[i][j] // 5)
                        count += 1
                
                # # 확산된 먼지 개수 만큼 제거
                new[i][j] += (graph[i][j] - (graph[i][j] // 5) * count)

    graph = new

# 공기청정기 위쪽 작동
def air_Up(airClearner):
    # 첫번째 거에서는 반시계
    x, y = airClearner[0]
    nx, ny = x, y

    dx = [0, -1, 0, 1]
    dy = [1, 0, -1, 0]

    dir = 0
    before = 0

    while True:
        # 벽 끝까지 간 경우
        if not (0 <= nx + dx[dir] < r and 0 <= ny + dy[dir] < c):
            dir += 1
        
        nx = nx + dx[dir]
        ny = ny + dy[dir]
        # 공기청정기인 경우
        if nx == x and ny == y:
            break

        if 0 <= nx < r and 0 <= ny < c:
            graph[nx][ny],before = before, graph[nx][ny]

for i in range(r):
    if graph[i][0] == -1:
        airclearner.append((i, 0))

def air_down(airclearner):
    x, y = airclearner[1]
    nx, ny = x, y

    dx = [0, 1, 0, -1]
    dy = [1, 0, -1, 0]

    dir = 0
    before = 0

    while True:
        if not (0 <= nx + dx[dir] < r and 0 <= ny + dy[dir] < c):
            dir += 1
        
        nx = nx + dx[dir]
        ny = ny + dy[dir]

        if nx == x and ny == y:
            break
        
        graph[nx][ny], before = before, graph[nx][ny]

for _ in range(t):
    expand()
    air_Up(airclearner)
    air_down(airclearner)

answer = 0
answer = sum(graph)

print(answer + 2)
```
